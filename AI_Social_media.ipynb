# Brinary converstion

```python
#converting the categorical variable into nimericals
varlist = ['Female','Male']

#Defing the map function
def binary_map(x):
    return x.map({'yes':1, 'no':0})

#Applying the function to the media list
social[varlist] = social[varlist].apply(binary_map)

#checking the social media dataframe now
social
```

# Outliner detection

```python
def detect_outliers_zscore(data):
    outliers = []
    thres = 3
    mean = np.mean(data)
    std = np.std(data)
    #print(mean, std)
    for i in data:
        #print(i)
        z_score = (i-mean)/std
        if (np.abs(z_score) > thres):
            outliers.append(i)
    return outliers# Driver code
```

```python
sample_outliers_User_ID = detect_outliers_zscore(social['User ID'])
sample_outliers_EstimatedSalary= detect_outliers_zscore(social['EstimatedSalary'])
sample_outliers_Purchased = detect_outliers_zscore(social['Purchased'])
print("Outliers in User ID column from Z-scores method: ", sample_outliers_User_ID)
print("Outliers in EstimatedSalary column from Z-scores method: ", sample_outliers_EstimatedSalary)
print("Outliers in Purchased column from Z-scores method: ", sample_outliers_Purchased )
```

```python
from scipy.stats import zscore
z1 = np.abs(zscore(social['User ID']))
z2 = np.abs(zscore(social['EstimatedSalary']))
z3 = np.abs(zscore(social['Purchased']))
social['Zscore_User ID'] = z1
social['Zscore_EstimatedSalary'] = z2
social['Zscore_Purchased'] = z3
outliers_User_ID=len(social[social['Zscore_User ID']>3])
outliers_EstimatedSalary=len(social[social['Zscore_EstimatedSalary']>3])
outliers_Purchased=len(social[social['Zscore_Purchased']>3])
data_z = social[social['Zscore_User ID']<3]
data_z = social[social['Zscore_EstimatedSalary']<3]
data_z = social[social['Zscore_Purchased']<3]
social=data_z.drop(['Zscore_User ID', 'Zscore_EstimatedSalary', 'Zscore_Purchased'], axis=1)
social
```
