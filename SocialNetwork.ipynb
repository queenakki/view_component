# Binary converstion

```python
#converting the categorical variable into nimericals
varlist = ['Gender']

#Defing the map function
def binary_map(x):
    return x.map({'Female':1, 'Male':0})

#Applying the function to the media list
social[varlist] = social[varlist].apply(binary_map)

#checking the social media dataframe now
social
```

# Categorical converstion

```python
# Dropping the first column from status dataset
status = pd.get_dummies(social['Social Media'], drop_first = True)

# Adding the status to the original social dataframe
social = pd.concat([social, status], axis = 1)

# Dropping 'Social Media' as we have created the dummies for it
social.drop(['Social Media'], axis = 1, inplace = True)

social
```

# Outliner direction

```python
def detect_outliers_zscore(data):
    outliers = []
    thres = 3
    mean = np.mean(data)
    std = np.std(data)
    #print(mean, std)
    for i in data:
        #print(i)
        z_score = (i-mean)/std
        if (np.abs(z_score) > thres):
            outliers.append(i)
    return outliers# Driver code
```

```python
sample_outliers_Age = detect_outliers_zscore(social['Age'])
sample_outliers_Gender = detect_outliers_zscore(social['Gender'])
sample_outliers_EstimatedSalary= detect_outliers_zscore(social['EstimatedSalary'])
sample_outliers_Purchased = detect_outliers_zscore(social['Purchased'])
print("Outliers in Age column from Z-scores method: ", sample_outliers_Age)
print("Outliers in Gender column from Z-scores method: ", sample_outliers_Gender)
print("Outliers in EstimatedSalary column from Z-scores method: ", sample_outliers_EstimatedSalary)
print("Outliers in Purchased column from Z-scores method: ", sample_outliers_Purchased )
```
